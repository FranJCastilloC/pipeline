```python
import pandas as pd
from scraping import main
import datetime
import re
import pandas as pd
import numpy as np
import re
```

```python
start_date = datetime.date.today().strftime('%Y-%m-%d')
end_date = datetime.date.today().strftime('%Y-%m-%d')
datos = main(start_date, end_date)
```

```python
for key in datos.keys():
    print(key)
```

```python
# Extraer los reportes nicos
dic_organizado = {}
unique_repot = set()  # set para que almacene valores nicos
for key in datos.keys():  # datos.keys para que solo itere sobre los ttulos del diccionario
    base_name = "_".join(key.split("_")[:-1])  # key.split("_") divide el nombre del ttulo cada vez que encuentre "_"
                                              # [:-1] -> Toma todos los elementos menos el ltimo as nos deshacemos de la fecha
                                              # "_".join une nuevamente los elementos con un "_"
    unique_repot.add(base_name)

# Procesamos cada tipo de reporte nico
for repote in unique_repot:  
    dfs_list = []
    for key in datos.keys():
        if key.startswith(repote):  # Revisar si en datos key el nombre empieza con alguno de los valores de la lista de unique_repot
            date_str = key.split("_")[-1]  # Si se cumple la condicin extrae el ltimo valor dividido por "_" que es la fecha
            datos_copy = datos[key].copy()  # Creamos una copia de datos para no daar el original
            datos_copy['fecha'] = date_str  # Agregamos la columna fecha
            dfs_list.append(datos_copy)
    if dfs_list:
        dic_organizado[repote] = pd.concat(dfs_list, ignore_index=True)  # Combinamos todos los DataFrames del mismo tipo
```

```python
# Ya tienes los keys como lista
keys = list(dic_organizado.keys())  # Convierte las claves del diccionario en una lista para poder iterar sobre ellas

for key in keys:  # Recorre cada clave en la lista de claves
   # Crea un nombre de variable limpio eliminando "BB_" y reemplazando caracteres problemticos
   # Esto transformar por ejemplo "BB_BB_RFMPOperDia" en "RFMPOperDia"
   df_name = key.replace("BB_", "").replace("-", "_").replace(" ", "_")
   
   # Obtiene el DataFrame correspondiente del diccionario dic_organizado
   # usando la clave original
   df = dic_organizado[key]
   
   # Crea una variable global con el nombre limpio y le asigna el DataFrame
   # globals() es un diccionario que contiene todas las variables globales
   # Al usar globals()[df_name] = df, estamos creando dinmicamente una variable
   # con el nombre almacenado en df_name y asignndole el valor df
   globals()[df_name] = df    
   
   # Imprime un mensaje confirmando la creacin del DataFrame
   # Esto es til para verificar que cada DataFrame se ha creado correctamente
   print(f"Creado DataFrame {df_name}")
```
